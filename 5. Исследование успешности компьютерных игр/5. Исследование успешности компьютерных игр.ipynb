Это очень большие проценты, с которыми мы не можем ничего сделать в данном случае. Мы могли бы заменять пропуски для конкретной игры средней оценкой её же на других платформах, однако, во-первых, далеко не все игры выпускались на разных платформах, во-вторых, успех игры на разных платформах может быть совершенно разным, поэтому подобной заменой мы можем сильно исказить данные.  

Нам придётся работать с той небольшой частью данных, что у нас есть, а для удобства работы со столбцами пропуски мы заменим на нули (сделаем "заглушку"), что будет означать, что оценки у этой игры по какой-либо причине не было. Нам нужно будет удалять эти "заглушки" при анализе данных.

```python
display(df[df['user_score'] == 'tbd'].shape)
display(df[df['critic_score'].isna()].shape)
df[df['user_score'].isna()].shape
```

```python
df['critic_score'] = df['critic_score'].fillna(0)
df['user_score'] = df['user_score'].fillna(0)
df['user_score'] = df['user_score'].replace('tbd', 0)
```

```python
display(df[df['user_score'] == 'tbd'].shape)
display(df[df['critic_score'].isna()].shape)
df[df['user_score'].isna()].shape
```

#### Столбец `rating` 

В столбце `rating` 40,48% (6676 строк) пропусков. Мы не можем заменить эти значения, поэтому вместо пропусков поставим значения 'Unknown', чтобы была возможность отдельно исследовать эту группу.

У некоторого количества игр рейтинг можно найти в версии для другой платформы, ведь жанр игры неизменен. Мы создадим словарь с известными нам жанрами игр, а затем заполним значения 'Unknown' теми, что удалось найти.

```python
df[df['rating'].isna()].shape
```

```python
# создаём словарь с жанрами игр
games_rating_dict = df.copy()
games_rating_dict = games_rating_dict.dropna(subset=['rating'])
games_rating_dict = games_rating_dict.set_index('name').to_dict()['rating']

# заменяем пропуски значением 'Unknown'
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df[df['rating'].isna()].shape
```

```python
#сброс индексов из-за отсутствия 659 строки в исходном датафрейме
df.reset_index(drop=True, inplace=True)
```

```python
# создание цикла, который будет заменять значения 'Unknown' 
# на значения из словаря games_rating_dict, если найдётся ключ
i = 0 # счётчик для удачных попыток
j = 0 # счётчик для неудачных попыток
for game in range(len(df['rating'])):
    if (df.loc[game, 'rating'] == 'Unknown'):
        try:
            df.loc[game, 'rating'] = games_rating_dict[df.loc[game, 'name']]
            i += 1
        except:
            j += 1
print('Ключей найдено:', i, 'Ключей не найдено:', j)
```

Так нам удалось спасти ещё 2,6% данных.

### Топ-5 популярных платформ по регионам

```python
# функция, которая принимает сводную таблицу и обрезает её как нужно
def top_5(income):
    
    # сумма продаж, не вошедших в топ-5
    income.tail().sum().values[0]
    
    # создание строки с остальными платформами
    income = income.rename(index={income.index[5]:'other'})
    
    # сумма продаж остальных платформ
    income = income.replace({income[income.columns[0]][5]:income.tail().sum().values[0]})
    
    # итоговая таблица
    income = income.head(6)
    
    return income
```

```python
# сводная таблица по платформам для игр в Северной Америке
na_top5_platforms = df_actual.pivot_table(
                    index='platform',
                    values='na_sales', 
                    aggfunc='sum').sort_values(
                    by='na_sales', ascending=False)
na_top5_platforms
```

```python
# сводная таблица по платформам для игр в Европе
eu_top5_platforms = df_actual.pivot_table(
                    index='platform',
                    values='eu_sales',
                    aggfunc='sum').sort_values(
                    by='eu_sales', ascending=False)

# применяем функцию top_5 к получившейся таблице
eu_top5_platforms = top_5(eu_top5_platforms)

eu_top5_platforms
```

```python
# сводная таблица по платформам для игр в Японии
jp_top5_platforms = df_actual.pivot_table(
                    index='platform',
                    values='jp_sales',
                    aggfunc='sum').sort_values(
                    by='jp_sales', ascending=False)

# применяем функцию top_5 к получившейся таблице
jp_top5_platforms = top_5(jp_top5_platforms)

jp_top5_platforms
```

### Влияет ли рейтинг ESRB на продажи в отдельном регионе?

**«E» («Everyone») — «Для всех»**: Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера.  
**«E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»**: Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови.  
**«T» («Teen») — «Подросткам»**: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.  
**«M» («Mature») — «Для взрослых»**: Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории.

```python
# сводная таблица по рейтингу игр в Северной Америке
na_esrb = df_actual.pivot_table(
          index='rating',
          values='na_sales',
          aggfunc='sum').sort_values(
          by='na_sales', ascending=False)

# сводная таблица по рейтингу игр в Европе
eu_esrb = df_actual.pivot_table(
          index='rating',
          values='eu_sales',
          aggfunc='sum').sort_values(
          by='eu_sales', ascending=False)

# сводная таблица по рейтингу игр в Японии
jp_esrb = df_actual.pivot_table(
          index='rating',
          values='jp_sales',
          aggfunc='sum').sort_values(
          by='jp_sales', ascending=False)

display(na_esrb, eu_esrb, jp_esrb)
```

```python
def graphics_esrb(income):
    
    if income.columns[0] == 'na_sales':
        _ = 'Северной Америке'
    if income.columns[0] == 'eu_sales':
        _ = 'Европе'
    if income.columns[0] == 'jp_sales':
        _ = 'Японии'
        
    fig = plt.figure(figsize=(10,9))
    ax1 = fig.add_subplot(1, 1, 1)
    plt.title('Продажи по рейтингам в {}\n'.format(_), size=12)
    ax1.axis('off')
    
    ax2 = fig.add_subplot(2, 2, 1)
    ax2.bar(income.index.values, income[income.columns[0]])
    
    plt.title('Продажи по рейтингам', size=12)
    plt.xlabel('Продаж, млн. шт.', size=12)
    plt.ylabel('Жанр', size=12)
    
    ax3 = fig.add_subplot(2, 2, 2)
    explode = (0.07, 0.07, 0.07, 0.07, 0.07)
    ax3.pie(income[income.columns[0]], labels = income.index.values, radius=1.1, autopct='%1.1f%%', pctdistance = 0.75, explode=explode, textprops={'fontsize': 10})
    
    plt.title('Продажи по рейтингам', size=12)
    
    plt.show()
```

## Проверка гипотез

### Гипотезы о средних пользовательских рейтингах платформ Xbox One и PC

H_0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
H_а: Средние пользовательские рейтинги платформ Xbox One и PC разные.  
alpha = 0.05

```python
# создаём копию датафрейма без "заглушек" в столбце 'user score'
df_actual_us = df_actual[df_actual['user_score'] != 0].copy()

sample_1 = df_actual_us[df_actual_us['platform'] == 'XOne']['user_score']
sample_2 = df_actual_us[df_actual_us['platform'] == 'PC']['user_score']

alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Гипотезы о средних пользовательских рейтингах жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»)

H_0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.  
H_а: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
alpha = 0.05

```python
sample_1 = df_actual_us[df_actual_us['genre'] == 'Action']['user_score']
sample_2 = df_actual_us[df_actual_us['genre'] == 'Sports']['user_score']

# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
